## 🔍 BƯỚC 3: Verify Environment

````python
from pathlib import Path

OUTPUT_DIR = '/content/drive/MyDrive/vicostone-autoresearch'

print("\n🔍 VERIFYING ENVIRONMENT...")

# Check OUTPUT_DIR
if not Path(OUTPUT_DIR).exists():
    print(f"❌ ERROR: OUTPUT_DIR not found!")
    print("   Run Cell 1 first to mount Google Drive")
else:
    print(f"✅ OUTPUT_DIR exists: {OUTPUT_DIR}")

# Check Python files
for f in ['vicostone_monitor.py', 'data_collector.py', 'autonomous_runner.py']:
    path = Path(OUTPUT_DIR) / f
    status = "✅" if path.exists() else "❌"
    print(f"{status} {f}")

# Check git
if not Path(OUTPUT_DIR, '.git').exists():
    print("⚠️ WARNING: Not a git repo")
else:
    print("✅ Git repo verified")
````
